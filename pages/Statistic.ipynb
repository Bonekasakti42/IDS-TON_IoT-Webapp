```python
import pandas as pd
import numpy as np
from datetime import datetime

# Jumlah data yang ingin disimulasikan
n_samples = 1000

np.random.seed(42)

data = []

for i in range(n_samples):

    # 90% normal traffic
    if np.random.rand() < 0.9:

        duration = np.random.uniform(
            0.000000,
            0.049943
        )

        src_bytes = np.random.uniform(
            0,
            130
        )

        dst_bytes = np.random.uniform(
            0,
            89
        )

        src_ip_bytes = np.random.uniform(
            48,
            415
        )

        dst_ip_bytes = np.random.uniform(
            0,
            134
        )

        label = "Normal"

    # 10% attack traffic
    else:

        duration = np.random.uniform(
            0.05,
            20
        )

        src_bytes = np.random.uniform(
            130,
            5000
        )

        dst_bytes = np.random.uniform(
            89,
            5000
        )

        src_ip_bytes = np.random.uniform(
            415,
            5000
        )

        dst_ip_bytes = np.random.uniform(
            134,
            5000
        )

        label = "Attack"

    data.append([
        datetime.now() + pd.Timedelta(seconds=i),
        duration,
        src_bytes,
        dst_bytes,
        src_ip_bytes,
        dst_ip_bytes,
        label
    ])

df_simulasi = pd.DataFrame(
    data,
    columns=[
        "timestamp",
        "duration",
        "src_bytes",
        "dst_bytes",
        "src_ip_bytes",
        "dst_ip_bytes",
        "label"
    ]
)

print(df_simulasi.head())

df_simulasi.to_csv(
    "simulasi_traffic.csv",
    index=False
)
```
